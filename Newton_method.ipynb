```python
import math as m
import pandas as pd
```

```python
x_0 = float(input("Initial guess:"))
e = float(eval(input("Error bound:")))
```

```python
def f(x):
    return x**6 - x - 1

def f_d(x):
    return 6*x**5-1
```

```python
def nm_approx(x_0, e):
    x_N = x_0
    x_n = x_0 + 100
    i = 0
    data = [[0, x_N, f(x_N), "No value"]]

    while abs(x_N - x_n) > e:
        x_n = x_N
        x_N = x_N - (f(x_N) / f_d(x_N))
        i += 1
        data.append([i, x_N, f(x_N), abs(x_N - x_n)])

    return pd.DataFrame(data)
```

```python
df = nm_approx(x_0, e)
df.rename(columns={0: "Iteration", 1: "x_n", 2: "f(x_n)", 3: "|x_n - x_n-1|"}, inplace=True)
df.set_index("Iteration", inplace= True)
df.head(10)
```
